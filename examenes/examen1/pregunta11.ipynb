```python
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt

random.seed(42)
```

# Ejercicio con N = 10,000

Para calcular la probabilidad teórica es de la siguiente manera:
1. Si lanzamos 2 dados hay 6x6 = 36 combinaciones diferentes.
2. 2 de esas combinaciones suman 11 (5,6 y 6,5)
3. Entonces, la probabilidad teórica es 2/36 =

```python
def calcular_probabilidad(N):
    dado1 = np.random.randint(1, 6+1, N)
    dado2 = np.random.randint(1, 6+1, N)
    total = dado1 + dado2

    resultados = pd.DataFrame({'lanzamiento': total})
    n_true = resultados[resultados['lanzamiento'] == 11].count()
    probabilidad = n_true / N

    teorica = 2/36

    plt.hist(resultados, bins=11)
    plt.title(f'Histograma de la suma de dos dados \nProbabilidad Montecarlo = {probabilidad.iloc[0]} \nProbabilidad teórica = {teorica}')
    plt.grid()
    plt.show()
    return

calcular_probabilidad(N=10000)
```

La probabilidad calculada con la simulación es muy acercada, sin embargo, no es igual.

Esta probabilidad simulada podría ser aceptable para no tener que calcular la teórica.

# Ejercicio con N = 100

```python
calcular_probabilidad(N=100)
```

La simulación a comenzado a perder precisión.

Podemos observar que ahora la probabilidad simulada subió a 0.08 e incluso el histograma ha perdido su forma.

Necesitamos incrementar N para lograr converger la probabilidad
