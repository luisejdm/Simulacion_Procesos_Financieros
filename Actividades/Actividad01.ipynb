# Obtener datos con yfinance

Obtén datos con yfinance desde 2022 de netflix y otras 4 empresas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
tickers = ['NFLX', 'NVDA', 'CTO', 'META', 'MELI']

closes = yf.download(tickers, start='2022-01-01')['Adj Close']
closes.head()
```

# Gráfica de los de cierre

```python
plt.figure(figsize=(12,8))
plt.plot(closes, label=closes.columns)
plt.title('Precios de cierre')
plt.legend(loc='best')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.grid()
```

# Precios promedio por acción

```python
precios_promedio = np.array([closes.mean()])
precios_promedio
```

# Obtén los días en que Netflix cotizo por encima de su precio promedio

```python
promedio_nflx = closes['NFLX'].mean()
closes[closes[['NFLX']] > promedio_nflx][['NFLX']]
```
