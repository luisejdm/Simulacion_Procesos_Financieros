```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calcular_integral(f, a, b, N):
    x = np.random.uniform(a, b, N)
    y = f(x)

    integral_aproximada = (b - a) * np.mean(y)
    return integral_aproximada


def ejemplo_funcion(x):
    return x**2  

a = 0
b = 1
n = [10, 100, 1000, 10000, 100000]
resultados = []
for i in range(len(n)):
    resultados.append(calcular_integral(ejemplo_funcion, a, b, n[i]))
resultados
```

```python
def circulo(x, y):
        return x**2 + y**2
    
def calcular_pi(N):
    a = -1
    b = 1

    x = np.random.uniform(a, b, N)
    y = np.random.uniform(a, b, N)

    puntos_dentro = circulo(x, y) < 1
    n_bajo_curva = np.sum(puntos_dentro)
    respuesta_integral = (b-a) * (b-a) * n_bajo_curva / N # (b-a)^2 porque es el area de un cuadrado
    
    fig, ax = plt.subplots(figsize=(6,6))
    ax.set_xlim(a, b)
    ax.set_ylim(a, b)
    ax.scatter(x[~puntos_dentro], y[~puntos_dentro], color='cadetblue', s=3, label='Puntos Fuera')
    ax.scatter(x[puntos_dentro], y[puntos_dentro], color='r', s=3, label='Puntos Dentro')
    ax.set_title(f'EstimaciÃ³n de pi: {respuesta_integral}              n = {N}')
    ax.grid()
    ax.legend(loc='upper right')
```

```python
n = [10, 100, 1000, 10000, 100000]

for i in range(len(n)):
    calcular_pi(n[i])
```
